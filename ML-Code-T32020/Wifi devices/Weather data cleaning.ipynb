```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("weatherdata.csv",sep=';')
data = pd.DataFrame(data)    # converting into the data frame
```

```python
data['year'] = data['Time'].str[:4]
```

```python
data['Date'] = data['Time'].str[:10]
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data = data.dropna()
```

```python
data = data.set_index('Date')
```

```python
data
```

```python
data = data[data['year']== '2019']
```

```python
data = data[['AirTemp [°C]','GustSpeed [km/h]']]
```
