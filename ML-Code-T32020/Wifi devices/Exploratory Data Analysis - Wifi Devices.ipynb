```python
cbd_list = ['Geelong Library','Malop and Yarra - NE cnr','Little Malop and James - CCTV/wifi cabinet',
           'Busport Project office','Ryrie /Union Inside','Ryrie /Union Outside','Ryrie/Moorabool Outside',
            'Civic Cark Park entrance CCTV cabinet','155 Mercer Foyer','LaPorcheta','Myer South side','City Hall']
regional_list = ['St Albans reserve rotunda','st helens south end playground','rippleside north end near gym',
                'Barwon Valley Activity Centre - Basketball Courts',
                 'Barwon Valley Activity Centre - South Car Park entrance','Barwon Valley Activity Centre – Entrance']
```

1. Each Device id has unique Device name
2. Some Devices names has high number of records than others -  This may be due to new devices are installed in recent times

```python
Cbd_Regional = []
for row in data['name']:
    if row in cbd_list :  
        Cbd_Regional.append('cbd')
    else:           
        Cbd_Regional.append('regional')
```

```python
data['Cbd_Regional'] = Cbd_Regional
```

```python
q1 = data['total'].quantile(0.25)
q3 = data['total'].quantile(0.75)
iqr = q3-q1
upper_boundary  = q3 + 1.5*iqr
lower_boundary = q1 - 1.5*iqr
```

```python
median = data['total'].median()
```

```python
data["total"] = np.where(data["total"] > upper_boundary, median, data['total'])
```

# Summarizing the data into daily data

```python
data = data.set_index('Date')
data = data.groupby('Cbd_Regional').resample('D').median()
```

```python
# the inverse of groupby, reset_index
data = data.reset_index()

# set again the timestamp as index
data = data.set_index("Date")
```
