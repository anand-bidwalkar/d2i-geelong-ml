```python
import pandas as pd
import numpy as np
```

```python
df_ac = pd.read_csv('Armstrong Creek.csv')

df_bel = pd.read_csv('Belmont.csv')

df_gw = pd.read_csv('Geelong West.csv')

df_geelong = pd.read_csv('Geelong.csv')

df_grove = pd.read_csv('Grovedale.csv')

df_herne = pd.read_csv('Herne Hill.csv')

df_high = pd.read_csv('Highton.csv')

df_lara = pd.read_csv('Lara.csv')

df_mh = pd.read_csv('Manifold Heights.csv')

df_mar = pd.read_csv('Marshall.csv')

df_md = pd.read_csv('Mount Duneed.csv')

df_new = pd.read_csv('Newtown.csv')

df_sg = pd.read_csv('South Geelong.csv')

df_wh = pd.read_csv('Wandana Heights.csv')

df_wp = pd.read_csv('Waurn Ponds.csv')

df_properties_1 = pd.concat([df_ac, df_bel, df_gw, df_geelong, df_grove, df_herne, df_high, df_lara, df_mh, df_mar, df_md, df_new, df_sg, df_wh, df_wp])
```

```python
df_bp = pd.read_csv('Bell Park.csv')

df_bh = pd.read_csv('Bell Post Hill.csv')

df_bw = pd.read_csv('Breakwater.csv')

df_corio = pd.read_csv('Corio.csv')

df_dc = pd.read_csv('Drumcondra.csv')

df_eg = pd.read_csv('East Geelong.csv')

df_hah = pd.read_csv('Hamlyn Heights.csv')

df_nwc = pd.read_csv('Newcomb.csv')

df_nl = pd.read_csv('Norlane.csv')

df_ns = pd.read_csv('North Shore.csv')

df_rs = pd.read_csv('Rippleside.csv')

df_st = pd.read_csv('St Albans Park.csv')

df_wt = pd.read_csv('Whittington.csv')

df_th = pd.read_csv('Thomson.csv')

df_ng = pd.read_csv('North Geelong.csv')

df_properties_2 = pd.concat([df_bp, df_bh, df_bw, df_corio, df_dc, df_eg, df_hah, df_nwc, df_nl, df_ns, df_rs, df_st, df_wt, df_th, df_ng])
```

```python
df_geelong_properties_corelogic = pd.concat([df_properties_1, df_properties_2])
df_geelong_properties_corelogic.head()
```

```python
df_geelong_properties_corelogic.shape
```

```python
df_geelong_properties_corelogic['Sell Price'] = df_geelong_properties_corelogic['Sell Price'].str.replace(',', '').str.replace('$', '')
```

```python
df_geelong_properties_corelogic['Auction'] = df_geelong_properties_corelogic['Auction'].fillna(0).astype(int)
df_geelong_properties_corelogic.head()
```

```python
df_public = pd.read_csv('Geelong Property Records.csv')
df_public.head()
```

```python
df_public = df_public.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
```

```python
df_geelong_properties_final = pd.concat([df_public, df_geelong_properties_corelogic])
```
