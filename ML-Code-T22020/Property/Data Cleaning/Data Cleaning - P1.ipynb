```python
import pandas as pd
```

```python
def clean_data(dataframe, suburb_name):
    sell_new = dataframe['sold'].str.split("in", n = 1, expand = True)
    dataframe['Sell Price'] = sell_new[0]
    dataframe['Sell Date'] = sell_new[1]
    auction = [ 1 if '(Auction)' in info[1] else 0 for info in sell_new[1].iteritems()]
    dataframe['Auction'] = auction
    dataframe['Sell Price'] = dataframe['Sell Price'].str.replace('Sold', '').str.replace('$', '')
    dataframe['Sell Date'] = dataframe['Sell Date'].str.replace('\(Auction\)', '')
    
    last_new = dataframe['last sold'].str.split("in", n = 1, expand = True)
    dataframe['Last Sell Price'] = last_new[0]
    dataframe['Last Sell Date'] = last_new[1]
    dataframe['Last Sell Price'] = dataframe['Last Sell Price'].str.replace('Last Sold', '').str.replace('$', '')
    dataframe['Last Sell Date'] = dataframe['Last Sell Date'].str.replace('\(Auction\)', '')
    
    rent_new = dataframe['rent'].str.split("in", n = 1, expand = True)
    dataframe['Rent Amount PW'] = rent_new[0]
    dataframe['Rent Date'] = rent_new[1]
    dataframe['Rent Amount PW'] = dataframe['Rent Amount PW'].str.replace('Rent', '').str.replace('$', '').str.replace('pw', '')
    
    land_new = dataframe['land_size'].str.split("|", n = 1, expand = True)
    
    if(len(land_new.columns) == 2):
        dataframe['Land Size SQM'] = land_new[0]
        dataframe['Building Size SQM'] = land_new[1]
    elif(len(land_new.columns) == 1):
        dataframe['Land Size SQM'] = land_new[0]
        dataframe['Building Size SQM'] = '0'
    else:
        dataframe['Land Size SQM'] = '0'
        dataframe['Building Size SQM'] = '0'
    dataframe['Land Size SQM'] = dataframe['Land Size SQM'].str.replace('Land size:', '').str.replace('sqm', '')
    dataframe['Building Size SQM'] = dataframe['Building Size SQM'].str.replace('Building size:', '').str.replace('sqm', '')
    
    dataframe['house_info'] = dataframe['house_info'].str.replace('House:', '').str.replace('Townhouse:', '')
    house_new = dataframe['house_info'].str.split(' ', n = 1, expand = True)
    
    bedrooms, toilets, cars = [], [], []

    for info in house_new[1].iteritems():
        splitted_info = info[1].strip().split('  ')
        if(len(splitted_info) == 3):
            bedrooms.append(splitted_info[0])
            toilets.append(splitted_info[1])
            cars.append(splitted_info[2])
        elif(len(splitted_info) == 2):
            bedrooms.append(splitted_info[0])
            toilets.append(splitted_info[1])
            cars.append(0)
        elif((len(splitted_info) == 1) and splitted_info[0] != ''):
            bedrooms.append(splitted_info[0])
            toilets.append(0)
            cars.append(0)
        else:
            bedrooms.append(0)
            toilets.append(0)
            cars.append(0)

    dataframe['Bedrooms'] = bedrooms
    dataframe['Toilets'] = toilets
    dataframe['Cars'] = cars

    dataframe['Suburb'] = suburb_name
    
    dataframe = dataframe.drop(['Unnamed: 0', 'sold', 'last sold', 'rent', 'house_info', 'land_size'], axis = 1)
    
    return dataframe
```

```python
df_geelong_clean = clean_data(df_geelong, 'Geelong')
df_geelong_clean.to_csv('Geelong_Clean.csv')
df_geelong_clean.head()
```
